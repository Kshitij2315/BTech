```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

# Set seaborn style
sns.set(style="whitegrid")

# 1. Sleep Duration Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Sleep Duration'], bins=20, kde=True, color='skyblue')
plt.title("Sleep Duration Distribution")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

```

```python
# 2. Average Sleep Duration by Occupation
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Occupation", y="Sleep Duration", estimator='mean', palette="coolwarm")
plt.xticks(rotation=45, ha="right")
plt.title("Average Sleep Duration by Occupation")
plt.ylabel("Avg Sleep Duration (hours)")
plt.tight_layout()
plt.show()
```

```python
# 5. Box Plot for Gender-wise Sleep Disorder Count
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Gender", y="Sleep Disorder", palette="pastel")
plt.title("Gender-wise Sleep Disorder Distribution")
plt.xlabel("Gender")
plt.ylabel("Sleep Disorder")
plt.tight_layout()
plt.show()
```

```python
# Pie Chart for Sleep Disorder Distribution
plt.figure(figsize=(7, 7))
sleep_disorder_counts = df['Sleep Disorder'].value_counts()
colors = sns.color_palette('pastel')[0:len(sleep_disorder_counts)]

plt.pie(sleep_disorder_counts, labels=sleep_disorder_counts.index, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title("Sleep Disorder Distribution")
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot for Sleep Duration vs Stress Level
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Stress Level", y="Sleep Duration", hue="Gender", palette="Set1", s=80, edgecolor="black")
plt.title("Sleep Duration vs Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("Sleep Duration (hours)")
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```
